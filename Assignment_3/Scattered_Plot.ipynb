### 1) Plot a scatter plot for x = [5,7,8,7,2,17,2,9,4,11,12,9,6] and y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

```python
import matplotlib.pyplot as plt
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y)
plt.title("Scattered Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### 2) Plot a scatter plot for log(x) for x= 1 to 10

```python
import numpy as np
x1 = [ i for i in range(1,11)]
y1 = np.log10(x1)
plt.scatter(x1, y1)
plt.title("log(x)")
plt.xlabel("x")
plt.ylabel("y=log10(x)")
plt.show()
```
